```python
 # =========================================Read Image==========================================

# Reading image BGR
Symbol_Binary=cv.imread('./cells/131520.jpg', cv.IMREAD_GRAYSCALE)

# print(np.shape(Symbol_Binary))
# Show Images
show_images([Symbol_Binary],titles=['Gray'])
print(np.shape(Symbol_Binary))
```

```python
hor_kernel = cv.getStructuringElement(cv.MORPH_RECT, (10, 1))
ver_kernel = cv.getStructuringElement(cv.MORPH_RECT, (1, 10))


Hor_Openned_img=cv.morphologyEx(np.copy(Symbol_Binary), cv.MORPH_OPEN, hor_kernel,iterations=5)
Ver_Openned_img=cv.morphologyEx(np.copy(Symbol_Binary), cv.MORPH_OPEN, ver_kernel,iterations=5)

show_images([Hor_Openned_img,Ver_Openned_img],titles=['Hor','Ver'])
```

```python
from skimage.transform import (hough_line, hough_line_peaks)
image=np.copy(Hor_Openned_img)
tested_angles = np.linspace(-np.pi / 2, np.pi / 2, 180)

# Perform Hough Transformation to change x, y, to h, theta, dist space.
hspace, theta, dist = hough_line(image, tested_angles)

plt.figure(figsize=(10,10))
plt.imshow(hspace)  


#Now, to find the location of peaks in the hough space we can use hough_line_peaks
h, q, d = hough_line_peaks(hspace, theta, dist)


angle_list=[]  #Create an empty list to capture all angles

# Generating figure 1
fig, axes = plt.subplots(1, 3, figsize=(15, 6))
ax = axes.ravel()

ax[0].imshow(image, cmap='gray')
ax[0].set_title('Input image')
ax[0].set_axis_off()

ax[1].imshow(np.log(1 + hspace),
             extent=[np.rad2deg(theta[-1]), np.rad2deg(theta[0]), dist[-1], dist[0]],
             cmap='gray', aspect=1/1.5)
ax[1].set_title('Hough transform')
ax[1].set_xlabel('Angles (degrees)')
ax[1].set_ylabel('Distance (pixels)')
ax[1].axis('image')

ax[2].imshow(image, cmap='gray')

origin = np.array((0, image.shape[1]))

for _, angle, dist in zip(*hough_line_peaks(hspace, theta, dist)):
    angle_list.append(angle) #Not for plotting but later calculation of angles
    y0, y1 = (dist - origin * np.cos(angle)) / np.sin(angle)
    ax[2].plot(origin, (y0, y1), '-r')
ax[2].set_xlim(origin)
ax[2].set_ylim((image.shape[0], 0))
ax[2].set_axis_off()
ax[2].set_title('Detected lines')

plt.tight_layout()
plt.show()

###############################################################
# Convert angles from radians to degrees (1 rad = 180/pi degrees)
angles = [a*180/np.pi for a in angle_list]

# Compute difference between the two lines
angle_difference = np.max(angles) - np.min(angles)
print(180 - angle_difference)   #Subtracting from 180 to show it as the small angle between two lines

# temp=np.copy(Hor_Openned_img)
# houghSpace,angles, dists = hough_line(temp)
# print(np.shape(houghSpace))
# print(len(angles))
# print(len(dists))

# show_images([temp],titles=['temp'])

# houghSpace,angles, dists = hough_line_peaks(houghSpace,angles, dists, threshold=0.45*np.max(houghSpace))
# print(np.shape(houghSpace))
# print(len(angles))
# print(len(dists))
# print(houghSpace)

# ax2.imshow(Ver_Openned_img, cmap=cm.gray)
# row1, col1 = image.shape
# for _, angle, dist in zip(*hough_line_peaks(houghSpace,angles, dists)):
#     y0 = (dist - 0 * np.cos(angle)) / np.sin(angle)
#     y1 = (dist - col1 * np.cos(angle)) / np.sin(angle)
#     ax2.plot((0, col1), (y0, y1), '-r')
# ax2.axis((0, col1, row1, 0))
# ax2.set_title('Detected lines')
# ax2.set_axis_off()


# print(angles)
# lines = cv.HoughLines(np.copy(Hor_Openned_img),1,np.pi/180,200)
# print(lines)
# for line in lines:
#     rho,theta = line[0]
#     a = np.cos(theta)
#     b = np.sin(theta)
#     x0 = a*rho
#     y0 = b*rho
#     x1 = int(x0 + 1000*(-b))
#     y1 = int(y0 + 1000*(a))
#     x2 = int(x0 - 1000*(-b))
#     y2 = int(y0 - 1000*(a))
    
#     cv.line(Hor_Openned_img,(x1,y1),(x2,y2),50,2)
# houghSpace,angles, dists = hough_line_peaks(houghSpace,angles, dists, threshold=0.45*np.max(houghSpace))
```

```python
 # =========================================Read Image==========================================

# Reading image BGR
Symbol_Binary=cv.imread('./cells/111010.jpg', cv.IMREAD_GRAYSCALE)

# print(np.shape(Symbol_Binary))
# Show Images
show_images([Symbol_Binary],titles=['Gray'])
print(np.shape(Symbol_Binary))

print(Symbol_Binary)
contours,hierarchy = cv.findContours(np.copy(Symbol_Binary), cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
print("Number of contours detected:", len(contours))


image_with_contours=np.copy(Symbol_Binary)
cv.drawContours(image_with_contours,contours, -1, 150, 2)
show_images([image_with_contours],titles=['Contours'])

image_with_contours_2=np.copy(Symbol_Binary)

for cnt in contours:
    x1,y1 = cnt[0][0]
    approx = cv.approxPolyDP(cnt, 0.01*cv.arcLength(cnt, True), True)
    if len(approx) == 4:
        x, y, w, h = cv.boundingRect(cnt)
        ratio = float(w)/h
        if ratio >= 0.9 and ratio <= 1.1:
            image_with_contours_2 = cv.drawContours(image_with_contours_2, [cnt], -1, 150, 3)
            cv.putText(image_with_contours_2, 'Square', (x1, y1), cv.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 0), 2)
        else:
            cv.putText(image_with_contours_2, 'Rectangle', (x1, y1), cv.FONT_HERSHEY_SIMPLEX, 0.6,150, 2)
            image_with_contours_2 = cv.drawContours(image_with_contours_2, [cnt], -1, (0,255,0), 3)

show_images([image_with_contours_2],titles=['Contours'])
```

```python
#Thiing image
vertical_kernel = cv.getStructuringElement(cv.MORPH_RECT, (5,1))
                                        
print(vertical_kernel)
eroded_image = cv.erode(Symbol_Binary, vertical_kernel, iterations=4)
show_images([eroded_image],titles=['Erroded'])
```

```python
# Use canny edge detection
edges = cv.Canny(eroded_image,50,150,apertureSize=3)
show_images([edges],titles=['Edges'])


#  Apply HoughLinesP method to
# to directly obtain line end points
lines_list =[]
lines =cv.HoughLines(np.copy(edges), 10, np.pi / 180, 150, None, 0, 0)

print(lines)

Symbol_Hough=np.copy(Symbol_Binary)

if lines is not None:
    for i in range(0, len(lines)):
        rho = lines[i][0][0]
        theta = lines[i][0][1]
        a = math.cos(theta)
        b = math.sin(theta)
        x0 = a * rho
        y0 = b * rho
        pt1 = (int(x0 + 1000*(-b)), int(y0 + 1000*(a)))
        pt2 = (int(x0 - 1000*(-b)), int(y0 - 1000*(a)))
        cv.line(Symbol_Hough, pt1, pt2, 50,2, cv.LINE_AA)
show_images([Symbol_Hough],titles=['Symbol_Hough'])
     
# # Save the result image
# cv2.imwrite('detectedLines.png',image)
```
